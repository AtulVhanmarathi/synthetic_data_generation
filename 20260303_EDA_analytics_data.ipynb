```python
import pandas as pd

def infer_and_set_dtypes(df):
    df = df.copy()
    
    for col in df.columns:
        # Skip if already numeric or datetime
        if pd.api.types.is_numeric_dtype(df[col]) or pd.api.types.is_datetime64_any_dtype(df[col]):
            continue
        
        # Try datetime conversion
        try:
            converted = pd.to_datetime(df[col], errors='raise')
            df[col] = converted
            continue
        except:
            pass
        
        # Try numeric conversion
        try:
            converted = pd.to_numeric(df[col], errors='raise')
            
            # Check if integer
            if (converted.dropna() % 1 == 0).all():
                df[col] = converted.astype("Int64")  # nullable integer
            else:
                df[col] = converted.astype(float)
                
        except:
            # Leave as string
            df[col] = df[col].astype("string")
    
    return df
```

```python
fact_maintenance_job = infer_and_set_dtypes(fact_maintenance_job)
```

```python
def create_month_column(df, datetime_col, new_col="month", format_type="yyyymm"):
    
    df = df.copy()
    df[datetime_col] = pd.to_datetime(df[datetime_col], errors="coerce")
    
    if format_type == "yyyymm":
        df[new_col] = df[datetime_col].dt.year * 100 + df[datetime_col].dt.month
        
    elif format_type == "period":
        df[new_col] = df[datetime_col].dt.to_period("M")
        
    elif format_type == "string":
        df[new_col] = df[datetime_col].dt.strftime("%Y%m")
        
    else:
        raise ValueError("format_type must be 'yyyymm', 'period', or 'string'")
        
    return df
```

```python
fact_maintenance_job = create_month_column(fact_maintenance_job, "close_datetime", "close_month", "period")
```
